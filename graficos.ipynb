```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
stocks=[
    {
        'ticker': 'AAPL',
        'name': 'Apple'
    },
    {
        'ticker': 'MSFT',
        'name': 'Microsoft'
    },
    {
        'ticker': 'DIS',
        'name': 'Disney'
    }
]
```

```python
def create_plots(stocks):
    data_disney_x=['2020-05-20','2020-06-20','2020-07-20','2020-08-20','2020-09-20','2020-10-20']
    data_disney_y=[100, 110, 105, 120, 140, 135]

    for stock in stocks:
        plt.figure(figsize=(10,6))
        plt.plot(data_disney_x, label=stock['name'])
        plt.legend()
        plt.ylabel('Returns')
        plt.xlabel('Time')
        plt.show()
    
```

```python
create_plots(stocks )
```
